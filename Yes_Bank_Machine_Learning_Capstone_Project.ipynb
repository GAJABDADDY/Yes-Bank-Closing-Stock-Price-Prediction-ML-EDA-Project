###**3.1. Missing Values:**

```python
# Visualizing the missing values
sns.heatmap(df.isnull(), cbar=False)
```

As we can see from the above graph, the data is perectly free of any missing values. If there had been any missing values, white sripes would have appeared.

Thus, we can say that there is'nt any need for Data Cleaning.

We don't need to use any imputation methods either as there are no missing values present.

Although there isn't any need for Data Cleaning, we can see that the column **'Date'** isn't properly formatted.

Thus, as a part of Data Transformation, we'll convert **'Date'** column's datatype to **Datetime** format.

```python
# changing 'date' column's datatype to datetime

df['Date'] =pd.to_datetime(df['Date'].apply(lambda x: datetime.strptime(x,'%b-%y')))
```
